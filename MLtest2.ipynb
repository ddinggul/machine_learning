```python
#데이터셋 불러오기
from sklearn import datasets
iris = datasets.load_iris()
print(iris)
```

```python
#특징량 & 분류 이름 확인
print("특징량 이름 = ", iris.feature_names)
print("분류 이름 = ", iris.target_names)
print("분류 값 = ", iris.target)
```

```python
# 데이터를 데이터 프레임으로 변환
import pandas as pd
df = pd.DataFrame(iris.data)
df.head() # head = 위에서 5줄 표시
```

```python
#1행에 열 이름 설정
df.columns = iris.feature_names
df["target"] = iris.target
df.head()
```

```python
# 히스토그램으로 나타내기
%matplotlib inline
import matplotlib.pyplot as plt

# 세 가지 품종을 별개의 데이터프레임으로 구분
# target의 값이 0이라면 df0에, 1이라면 df1에, 2라면 df2에 넣는다
df0 = df[df["target"] == 0]
df1 = df[df["target"] == 1]
df2 = df[df["target"] == 2]

#세 가지의 품종을 히스토그램에서 색으로 구분해 그린다.
#꽃받침의 폭으로 히스토그램을 그린다.
plt.figure(figsize = (5,5))
xx = "sepal width (cm)" # sepal width(cm) 열을 대상으로
df0[xx].hist(color="b", alpha = 0.5) # 파란색 히스토그램
df1[xx].hist(color="r", alpha = 0.5) # 빨간색 히스토그램
df2[xx].hist(color="g", alpha = 0.5) # 초록색 히스토그램
plt.xlabel(xx) # 그래프에 x축 라벨을 작성
plt.show()
```

```python
#산포도로 그리기 - 2가지 특징 사용
xx = "sepal width (cm)" # sepal width(cm) 열을 첫 번째 대상으로
yy = "sepal length (cm)" #sepal length(cm) 열을 두 번째 대상으로
plt.figure(figsize=(5,5))
plt.scatter(df0[xx], df0[yy], color="b", alpha = 0.5) #파란색 산포도
plt.scatter(df1[xx], df1[yy], color="r", alpha = 0.5) #빨간색 산포도
plt.scatter(df2[xx], df2[yy], color="g", alpha = 0.5) #초록색 산포도
plt.xlabel(xx) # 그래프에 x라벨 생성
plt.ylabel(yy) # 그래프에 y라벨 생성
plt.grid()
plt.show()
```

```python
# 3D 산포도 그리기
from mpl_toolkits.mplot3d import Axes3D # Axes3D 임포트
#꽃받침 폭, 꽃받침 길이, 꽃잎 길이로 3D 산포도로 그린다
xx = "sepal width (cm)"
yy = "sepal length (cm)"
zz = "petal length (cm)"
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(projection = "3d")
ax.scatter(df0[xx], df0[yy], df0[zz], color = "b")
ax.scatter(df1[xx], df1[yy], df1[zz], color = "r")
ax.scatter(df2[xx], df2[yy], df2[zz], color = "g")
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
plt.show()
```

```python
#시점 변경 - 1
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(projection = "3d")
ax.scatter(df0[xx], df0[yy], df0[zz], color = "b")
ax.scatter(df1[xx], df1[yy], df1[zz], color = "r")
ax.scatter(df2[xx], df2[yy], df2[zz], color = "g")
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
ax.view_init(0,240) # 시점 변경 코드
plt.show()
```

```python
#시점 변경 - 2
fig = plt.figure(figsize = (5,5))
ax = fig.add_subplot(projection = "3d")
ax.scatter(df0[xx], df0[yy], df0[zz], color = "b")
ax.scatter(df1[xx], df1[yy], df1[zz], color = "r")
ax.scatter(df2[xx], df2[yy], df2[zz], color = "g")
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
ax.view_init(90,270) # 시점 변경 코드
plt.show()
```
